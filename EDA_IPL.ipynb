```python
def scaling(batsman):
  return (batsman - batsman.min()) / (batsman.max() - batsman.min())
```

```python
df.innings_played = scaling(df.innings_played)
df.previous_average = scaling(df.previous_average)
df.previous_strike_rate= scaling(df.previous_strike_rate)
df.runs = scaling(df.runs)
```
